Exercise 1: Factor List
Write a function that takes an integer as input and returns a list containing all the factors of that number.

Example:
Input: 12
Output: [1, 2, 3, 4, 6, 12]

```python
x=int(input("Enter the number to get a factor list"))
j=1
y = []
for j in range(1,x+1):
    if (x%j)==0:
        y.append(j)
print(f"The output is {y}")
    
```

Exercise 2: Vowel Count
Write a function that takes a string as input and returns a list containing the count of each vowel (a, e, i, o, u) in the input string.

Example:
Input: "Hello, World!"
Output: [0, 1, 0, 2, 0]

```python
import numpy as np
h= input("Enter the string: ")
vowel = ["a","e","i","o","u"]
h = np.array(list(h))
print(h)
store = []
for x in vowel:
    count=0
    for j in h:
        if x==j:
            count+=1
    store.append(count) 
print(f"The vowels(a,e,i,o,u) in the string is in the order {store}")
    
```

Exercise 3: Word Lengths
Write a function that takes a string as input and returns a list containing the length of each word in the input string.

Example:
Input: "This is a sample sentence"
Output: [4, 2, 1, 6, 8]

```python
import numpy as np
h= input("Enter the string")
h = h.split()
store = []
for x in h:
    store.append(len(x))
print(f"Lengths of each word in string {h} is {store}")
```

Exercise 5: Palindrome Check
Write a function that takes a string as input and returns a list containing True or False for each word in the input string, indicating whether the word is a palindrome or not.

Example:
Input: "radar level hello"
Output: [True, True, False]

```python
def pallendrom_check(sentence):
    store = []
    for i in sentence:
        p = np.array(list(i))
        count = (len(p)+1)/2
        run = len(p)
        j = 0
        is_palindrome = all(q == p[run-1-j] for j, q in enumerate(p))
        if is_palindrome:
            store.append("True")
        else:
            store.append("False")    
    return (store)  

input_string = input("Sentence to check pallendrome")
input_line = input_string.split()
value = pallendrom_check(input_line)
print(f"Indications of {input_string} of pallendrom check is : {value}")
```

Exercise 6: Integer to Digits
Write a function that takes an integer as input and returns a list containing its individual digits.

Example:
Input: 4567
Output: [4, 5, 6, 7]

```python
def convert_digits(value):
    store = []
    p = np.array(list(value))
    for item in p:
        store.append(int(item))
    return store
number = input ("Enter the value : ")
s = convert_digits(number)
print(f"The values in {number} are {s}")
```
